# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

# plot

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x, y)

plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('title\nsubtitle')
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [2,3,4,5]
x2 = [1,2,3,4]
y2 = [7,8,9,10]

plt.plot(x1, y1, label='cost')
plt.plot(x2, y2, label='income')

plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
y = x**2 + 2

plt.plot(x, y, color='green', linestyle='-.', linewidth=2)
plt.grid()
plt.show()
```

```python
plt.plot(x, y, color='green', marker='*', markersize=5, label='graph')
plt.legend()
plt.grid()
plt.show()
```

# scatter

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,3,5,4,6,8,7,7,10]

plt.scatter(x, y, label='scatter', s=60, color='r', marker='*')
plt.legend()
plt.show()
```

# bar

```python
x = [1399,1400,1401,1402]
y = [1000,1500,1263,1100]

plt.bar(x, y, label='sale', color='b',width=0.4)

plt.legend()
plt.show()
```

# histogram

```python
x = np.random.normal(30,8,100)
# bins = [10,20,30,40,50,60]
plt.hist(x) # , bins=bins
plt.show()
```

# pie

```python
slices = [7,3,2,1]
activities = ['sleeping', 'eating', 'working', 'playing']

plt.pie(slices, labels=activities, colors=['red', 'blue', 'm', 'yellow'], explode=[0,0.1,0,0],
       autopct='%1.1f%%')

plt.show()
```
